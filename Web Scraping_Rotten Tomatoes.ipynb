```python
tomatometer_score = []
audience_score_rt = []
distributor = []
director = []
producer = []
genre = []
runtime = []
original_language = []
MPA_Ratings = []
```

```python
from datetime import datetime
print(datetime.now())
for i in df2['movie']: 
    # searching the movie on rt
    search_url = 'https://www.rottentomatoes.com/search?search='+str(i)
    response_search = requests.get(search_url)
    soup_search = BeautifulSoup(response_search.content, 'html.parser')

    # obtaining the movie page
    try:
        movie_url = soup_search.select('search-page-media-row a')[0].get('href')
        response = requests.get(movie_url)
        soup = BeautifulSoup(response.content, 'html.parser')
 
        # adding tomatometer score to list
        try:
            score_tm = soup.select("#topSection > div.thumbnail-scoreboard-wrap > score-board")[0]["tomatometerscore"]
            tomatometer_score.append(score_tm)
        except:
            tomatometer_score.append(np.nan) 
            
        # adding audience score to list
        try:
            score_aud = soup.select("#topSection > div.thumbnail-scoreboard-wrap > score-board")[0]["audiencescore"]
            audience_score_rt.append(score_aud)
        except:
            audience_score_rt.append(np.nan)
            
        # adding distributor to list
        try:
            dist = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(11) > div.meta-value")[0].get_text()
            dist2 = ' '.join(dist.replace('\n','').split())                  
            distributor.append(dist2)
        except:
            distributor.append(np.nan)
            
         # adding producers to list
        try:
            prod = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(5) > div.meta-value")[0].get_text()
            prod2 = ' '.join(prod.replace('\n','').split())
            producer.append(prod2)
        except:
            producer.append(np.nan)

             # adding directors to list
        try:
            dir = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(4) > div.meta-value")[0].get_text()
            dir2 = ' '.join(dir.replace('\n','').split())
            director.append(dir2)
        except:
            director.append(np.nan)

             # adding genres to list
        try:
            gen = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(2) > div.meta-value.genre")[0].get_text()
            gen2 = ' '.join(gen.replace('\n','').split())
            genre.append(gen2)
        except:
            genre.append(np.nan)

            # adding runtime to list
        try:
            run = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(10) > div.meta-value > time")[0].get_text()
            run2= ' '.join(run.replace('\n','').split())
            runtime.append(run2)
        except:
            runtime.append(np.nan)

            # adding original language to list
        try:
            ol = soup.select("#mainColumn > section.panel.panel-rt.panel-box.movie_info.media > div > div > ul > li:nth-child(3) > div.meta-value")[0].get_text()
            ol2= ' '.join(ol.replace('\n','').split())
            original_language.append(ol2)
        except:
            original_language.append(np.nan)

           # adding MPA_Ratings to list
        try:
            MPA_R = soup.select("#topSection > div.thumbnail-scoreboard-wrap > score-board")[0]["rating"]
            MPA_R2= ' '.join(MPA_R.replace('\n','').split()).partition('(')[0].strip()
            MPA_Ratings.append(MPA_R2)
        except:
            MPA_Ratings.append(np.nan)
                           
    except:
        tomatometer_score.append(np.nan)
        audience_score_rt.append(np.nan)
        distributor.append(np.nan)                      
        producer.append(np.nan)
        director.append(np.nan)
        genre.append(np.nan)
        runtime.append(np.nan)
        original_language.append(np.nan)
        MPA_Ratings.append(np.nan)

print(datetime.now())
```

```python
df_try = pd.DataFrame(list(zip(tomatometer_score, audience_score_rt, distributor, producer, director, genre, runtime, original_language, MPA_Ratings)),
               columns =['tomatometer_score', "audience_score_rt", "distributor", 'producer', "director", "genre", "runtime", "original_language", "MPA_Ratings"])              
```
