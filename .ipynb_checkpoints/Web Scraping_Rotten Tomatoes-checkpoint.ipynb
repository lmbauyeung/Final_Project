```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("The_Numbers_data_cleaned.csv")
df
```

```python
urls = []
for i in df["MOVIE"].to_list():
    urls.append('https://www.rottentomatoes.com/search?search='+str(i))
```

```python
urls
```

```python
re = requests.get(urls[0])
soup = BeautifulSoup(re.content, 'html.parser')
x = soup.select('search-page-media-row a')[0].get('href')
x
```

```python
list_urls = []
for u in range(0, 10):
    re = requests.get(urls[u])
    soup_search = BeautifulSoup(re.content, 'html.parser')
    list_urls.append(soup_search.select('search-page-media-row a')[0].get('href'))

    
```
